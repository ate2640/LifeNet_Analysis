### Individual LifeNet Analysis

The 35-49 & 50-64 age groups account for the largest recipients of organs with approximately 2/3 in both years, with 35-49
accounting for 21.23% & 23.75% in Years 1 and 2, respectively, and 50-64 accounting for 45.54% & 43.14% in Years 1 & 2,
respectively

African-Americans account for more than half of recipients in both years, and there was a 5% drop in African-American recipients from Year 1 to Year 2

Non-Whites accounted for the largest proportion of recipients, at roughly 69%

There was a notable increase in Hispanic recipients, from 6.15% to 9.37%

Men account for 3/5 of recipients in both years

There was a 5% increase in recipients of the 18-34 age group 

There was a 5.3% decrease in receipients of the Senior Citizen group
